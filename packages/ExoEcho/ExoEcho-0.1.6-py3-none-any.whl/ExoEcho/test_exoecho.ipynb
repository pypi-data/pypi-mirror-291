```python
import pandas as pd
from Functions import *
from Telescope import *
import Ariel

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
```

```python
jwstNirspec = Telescope(name="JWST NIRSpec 0.6-5.3um R=100", 
                        diameter=6.5, 
                        wavelength_range=(0.6, 5.3), 
                        resolution=100, 
                        throughput=0.36)

jwstNirspec.constructTable()
```

```python
# preparing figure
fig, ax = plt.subplots()

# plotting noise estimate for JWST NIRSpec at R=100
jwstNirspec.plotParam("TOI-2338b", "Noise Estimate", ax=ax, marker='.', label="Noise Estimate")

# plotting transmission flux for JWST NIRSpec at R=100
jwstNirspec.plotParam("TOI-2338b", "Transit Flux", ax=ax, marker='.', label="Transmission Flux")

# adding grid
ax.grid(which='major', linestyle='--', linewidth=0.5)

# showing legend
ax.legend()

# adding labels
ax.set_ylabel("Precision [ppm]")
ax.set_xlabel("Wavelength [$\mu m$]")

# adding title
ax.set_title("Noise Estimate for TOI-2338b\nusing JWST NIRSpec at R=100")

plt.show()
```
